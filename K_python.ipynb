```
from konlpy.tag import Okt
from konlpy.utils import pprint
from googletrans import Translator
from pandas import DataFrame
from konlpy.corpus import kolaw
from konlpy import utils
from konlpy.tag import Kkma
```

```
const = kolaw.open('constitution.txt').read() # carregar o corpus: Constituição sul coreana
const[:120]
```

```
#Dependência do Konlpy
kkma = Kkma()
#Detecção de sentnças
kkma.sentences(u'저는 대학생이구요. 소프트웨어 관련학과 입니다.')
```

```
#Analisar a frase para morfemas
kkma.morphs(u'자주 사용을 안하다보니 모두 까먹은 상태입니다.')
```

```
#Extrair substantivos
kkma.nouns(u'대한민국헌법 유구한 역사와 전통에 빛나는 우리')
```

```
# Realizando o part of speech tagging
okt = Okt()
kopos = okt.pos(const,norm=True,stem=True,join=True)
print(kopos)
```

```
Korean_words = DataFrame(kopos,columns=['original_word'])
Korean_words
```

```
analysis = Korean_words["original_word"].str.split("/")
Korean_words["korean"]= analysis.str[0]
Korean_words["classification"]= analysis.str[1]
Korean_words.head(10)
```
